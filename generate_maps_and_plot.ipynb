```python
# install wb_command with "sudo apt install connectome-workbench"
# download 'HCP_S1200_997_tfMRI_ALLTASKS_level2_cohensd_hp200_s2_MSMAll.dscalar.nii' from HCP database

# transform HCP tasks into a nifti file (-cifti-separate):
import subprocess
subprocess.check_output("wb_command -cifti-separate \
HCP_S1200_997_tfMRI_ALLTASKS_level2_cohensd_hp200_s2_MSMAll.dscalar.nii COLUMN -volume-all  \
tasks_nifti.nii", shell=True);

# separate each task into a different nifti file:
subprocess.check_output("wb_command -volume-merge \
language.nii -volume tasks_nifti.nii -subvolume 66", shell=True);
subprocess.check_output("wb_command -volume-merge \
WM.nii -volume tasks_nifti.nii -subvolume 11", shell=True);
subprocess.check_output("wb_command -volume-merge \
motorRH.nii -volume tasks_nifti.nii -subvolume 48", shell=True);
subprocess.check_output("wb_command -volume-merge \
motorLH.nii -volume tasks_nifti.nii -subvolume 46", shell=True);
subprocess.check_output("wb_command -volume-merge \
motorRF.nii -volume tasks_nifti.nii -subvolume 47", shell=True);
subprocess.check_output("wb_command -volume-merge \
motorLF.nii -volume tasks_nifti.nii -subvolume 45", shell=True);
subprocess.check_output("wb_command -volume-merge \
motorT.nii -volume tasks_nifti.nii -subvolume 49", shell=True);
subprocess.check_output("wb_command -volume-merge \
emotion.nii -volume tasks_nifti.nii -subvolume 83", shell=True);
subprocess.check_output("wb_command -volume-merge \
social.nii -volume tasks_nifti.nii -subvolume 75", shell=True);
```

```python
# plot task maps
from nilearn import plotting
cut_coords=[-90, -85, -80, -75, -70, -65, -60, -55, -50, -45, -40, -35]
display_mode='y'
threshold=0.5
plotting.plot_prob_atlas('motorRH.nii', title='right hand', threshold=threshold, display_mode=display_mode, cut_coords=cut_coords)
plotting.plot_prob_atlas('motorLH.nii', title='left hand', threshold=threshold, display_mode=display_mode, cut_coords=cut_coords)
plotting.plot_prob_atlas('motorRF.nii', title='right foot', threshold=threshold, display_mode=display_mode, cut_coords=cut_coords)
plotting.plot_prob_atlas('motorLF.nii', title='left foot', threshold=threshold, display_mode=display_mode, cut_coords=cut_coords)
plotting.plot_prob_atlas('motorT.nii', title='tongue', threshold=threshold, display_mode=display_mode, cut_coords=cut_coords)
plotting.plot_prob_atlas('language.nii', title='language', threshold=threshold, display_mode=display_mode, cut_coords=cut_coords)
plotting.plot_prob_atlas('WM.nii', title='working memory', threshold=threshold, display_mode=display_mode, cut_coords=cut_coords)
plotting.plot_prob_atlas('emotion.nii', title='emotion', threshold=threshold, display_mode=display_mode, cut_coords=cut_coords)
plotting.plot_prob_atlas('social.nii', title='social', threshold=threshold, display_mode=display_mode, cut_coords=cut_coords)
```
